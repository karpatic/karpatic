LineCurve
A curve representing a 2d line segment.
- LineCurve( startVec : Vector2, endVec : Vector2 )

CatmullRomCurve3
.curveType : String
Possible values are centripetal, chordal and catmullrom.
.tension : Float
When .curveType is catmullrom, defines catmullrom's tension.

SplineCurve
Create a smooth 2d spline curve from a series of points.
uses Interpolations.CatmullRom to create the curve.
- SplineCurve( points : Array )

ARC/ EllipseCurve
- Creates a 2d curve in the shape of an ellipse.

QuadraticBezierCurve
Create a smooth 2d quadratic bezier curve, defined by a startpoint, endpoint and a single control point.
QuadraticBezierCurve( v0 : Vector2, v1 : Vector2, v2 : Vector2 )

CubicBezierCurve
Create a smooth 2d cubic bezier curve, defined by a start point, endpoint and two control points.

Path

.currentPoint : Vector2
The current offset of the path. Any new Curve added will start here.

.splineThru ( points : Array ) : this
points - An array of Vector2s
Connects a new SplineCurve onto the path.

.setFromPoints ( vector2s : Array ) : this
points -- array of Vector2s.
Points are added to the curves array as LineCurves.

.lineTo ( x : Float, y : Float ) : this
Connects a LineCurve from .currentPoint to x, y onto the path.

.moveTo ( x : Float, y : Float ) : this
Move the .currentPoint to x, y.

.quadraticCurveTo ( cpX : Float, cpY : Float, x : Float, y : Float ) : this
Creates a quadratic curve from .currentPoint with cpX and cpY as control point and updates .currentPoint to x and y.

.bezierCurveTo ( cp1X : Float, cp1Y : Float, cp2X : Float, cp2Y : Float, x : Float, y : Float ) : this
This creates a bezier curve from .currentPoint with (cp1X, cp1Y) and (cp2X, cp2Y) as control points and updates .currentPoint to x and y.

<script type="module">

import * as THREE from "three";
window.THREE = THREE
let CSG = (await import('../js/CSG.js')).CSG
console.log({CSG})
let { create } = (await import('https://cdn.jsdelivr.net/npm/d3@7.8.4/dist/d3.min.js')); 
import { OrbitControls } from "three/addons/controls/OrbitControls.js";
import { SVGLoader } from "three/addons/loaders/SVGLoader.js";
let width = 500; let height = width;
// console.clear();

  let geomUrl="https://services1.arcgis.com/mVFRs7NF4iFitgbY/ArcGIS/rest/services/Tpop/FeatureServer/0/query?where=1%3D1&outFields=*&returnGeometry=true&f=pgeojson";
  let geojson = (await (await fetch(geomUrl) ).json())
  console.log({geojson})
  let path = d3.geoPath().projection(d3.geoMercator().fitExtent([[0, 0], [width, height]], geojson) )
  let svg = d3.create("svg")
  svg.selectAll("path")
     .data(geojson.features)
     .enter()
     .append("path")
     .attr("d", path)
     .attr("stroke", "#000")
     .attr("fill", "none")
     .attr("stroke-width", 0.5);
  svg = svg.node()
  console.log({'svg':svg.children});
  svg = svg.outerHTML


let scene = new THREE.Scene();
scene.background = new THREE.Color(0xface8d);
let camera = new THREE.PerspectiveCamera(60, innerWidth / innerHeight, 1, 5000);
camera.position.set(0, 0, 1000);
let renderer = new THREE.WebGLRenderer({ antialias: true });
renderer.setSize(innerWidth, innerHeight);
document.body.appendChild(renderer.domElement);
window.addEventListener("resize", (event) => {
  camera.aspect = innerWidth / innerHeight;
  camera.updateProjectionMatrix();
  renderer.setSize(innerWidth, innerHeight);
});

let controls = new OrbitControls(camera, renderer.domElement);
controls.enableDamping = true;

let light = new THREE.DirectionalLight(0xffffff, 1);
light.position.set(-1, 1, 1);
scene.add(light, new THREE.AmbientLight(0xffffff, 0.5));
let gu = { time: { value: 0 } };

let loader = new SVGLoader();
let data = loader.parse(svg);
console.log({data});
data.paths.forEach(path => {
  let pts = path.subPaths[0].getPoints(50);
  pts.pop();
  pts.forEach(p => p.y *= -1);
  pts.reverse();
  let strokeWidth = path.userData.style.strokeWidth;

  let pPrev = new THREE.Vector2(), pNext = new THREE.Vector2(), c = new THREE.Vector2();
  let offsetPtsOuter = [];
  let offsetPtsInner = [];
  pts.forEach((p, idx) => {
    let idxPrev = idx - 1;
    if (idxPrev < 0 ) idxPrev = pts.length - 1;
    let idxNext = idx + 1;
    if (idxNext== pts.length) idxNext = 0;
    pPrev.subVectors(pts[idxPrev], p).normalize();
    pNext.subVectors(pts[idxNext], p).normalize();
    let anglePrev = pPrev.angle();
    let angleNext = pNext.angle();
    let angleMid = (angleNext - anglePrev) * 0.5;
    pPrev.rotateAround(c, angleMid);
    let offsetDistOuter = 1 / Math.cos(angleMid - Math.PI * 0.5);
    let offsetDistInner = offsetDistOuter - strokeWidth;  // here you could adjust the thickness of the wall
    offsetPtsOuter.push(pPrev.clone().negate().setLength(offsetDistOuter).add(p));
    offsetPtsInner.push(pPrev.clone().negate().setLength(offsetDistInner).add(p));
  });

  let shapeOuter = new THREE.Shape(offsetPtsOuter);
  let shapeInner = new THREE.Shape(offsetPtsInner);

  let outerGeom = new THREE.ExtrudeGeometry(shapeOuter, {depth: 10, bevelEnabled: false,});
  let innerGeom = new THREE.ExtrudeGeometry(shapeInner, {depth: 10, bevelEnabled: false,});
  
  let material = new THREE.MeshLambertMaterial({color: "olive"});
  const outerGeomMesh = new THREE.Mesh(outerGeom, material); 
  const innerGeomMesh = new THREE.Mesh(outerGeom, material); 

  let outerBSP = CSG.fromMesh( outerGeomMesh )
  let innerBSP = CSG.fromMesh( innerGeomMesh )
  
  let bspResult = outerBSP.subtract(innerBSP);
  
  let oShape = CSG.toMesh( bspResult, outerBSP.matrix, material ); 
  // let oShape = new THREE.Mesh(resultGeom, material); 

  scene.add(oShape);
});


let clock = new THREE.Clock();

renderer.setAnimationLoop((_) => {
  let t = clock.getElapsedTime();
  gu.time.value = t;
  controls.update();
  renderer.render(scene, camera);
});
</script>
