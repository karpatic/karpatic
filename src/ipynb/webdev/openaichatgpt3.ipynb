```python
# import the  cv2_imshow() since the OpenCV imshow() functions cause the jupyter colab to crash.
import cv2
from urllib.request import urlopen
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Loading the cascade (pre-trained models + remote image)

cascades_path = 'drive/xxx/'
# Load the cascade
face_cascade = cv2.CascadeClassifier(cascades_path +'haarcascade_frontalface_default.xml')
```

```python
# Read the input image
req = urlopen('https://tvseriesfinale.com/wp-content/uploads/2014/04/siliconvalley03-590x331.jpg')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1) # 'Load it as it is' 

# spacefrom [topmost:fromTop, leftmost:fromLeft] 
img = img[0:256, 90:90+256]

h, w = img.shape[:2]
# Add an alpha channel, fully opaque (255) and save.
result_image = np.dstack((img.copy(), np.zeros((h,w),dtype=np.uint8)+255))
cv2.imwrite("./original.png", result_image)
```

```python
# Detect faces

# Convert frame to grayscale:
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# detectMultiScale  - Detects objects of different sizes in the input image.
# The detected objects are returned as a list of rectangles and its a part of face_cascade. 
# we are going to use scaleFactor of 1.1 and minNeighbors= 4 which is the default value.
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```python
# You can manually set a bounding box here if you want
```

```python
# draw a rectangle around the face and display the result

# Since we get the rectangle x,y with height and width from the previous step we are going to 
# loop around to draw for all the identified faces,

# Iterate over each detected face:
for (x, y, w, h) in faces:
        # Draw a rectangle  to see the detected face (debugging purposes):
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 0), -1)

# Display the resulting frame
cv2_imshow(img)
```

```python
from PIL import Image
import numpy as np

# Make into Numpy array of RGB and get dimensions
RGB = np.array(img)
h, w = RGB.shape[:2]

# Add an alpha channel, fully opaque (255)
RGBA = np.dstack((RGB, np.zeros((h,w),dtype=np.uint8)+255))

# Make mask of black pixels - mask is True where image is black
mBlack = (RGBA[:, :, 0:3] == [0,0,0]).all(2)

# Make all pixels matched by mask into transparent ones
RGBA[mBlack] = (0,0,0,0)

# Convert Numnpy array back to PIL Image and save
Image.fromarray(RGBA).save('mask.png')
cv2_imshow(RGBA)
```
