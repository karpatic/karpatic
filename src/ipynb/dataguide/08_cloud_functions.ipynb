# [ Quickstart ]( https://cloud.google.com/functions/docs/quickstart-python )

1) In the Cloud Console, on the project selector page, select or create a Cloud project.
[ GCloud Platform Project Selector ]( https://console.cloud.google.com/projectselector2/home/dashboard?_ga=2.178554084.1700393638.1583430572-875266049.1582662020&_gac=1.114841077.1582662020.Cj0KCQiAqNPyBRCjARIsAKA-WFzpKrN2CANBYiFYzQxDMN-_ckwddpHwC_QWPiGPgZFt_NjbX7FJxWkaAhReEALw_wcB )

2) Make sure that [ billing ]( https://cloud.google.com/billing/docs/how-to/modify-project ) is enabled for your Google Cloud project. 

3) Enable the Cloud Functions [ API ]( https://console.cloud.google.com/flows/enableapi?apiid=cloudfunctions&redirect=https://cloud.google.com/functions/docs/quickstart-python&_ga=2.254101704.1700393638.1583430572-875266049.1582662020&_gac=1.24998344.1582662020.Cj0KCQiAqNPyBRCjARIsAKA-WFzpKrN2CANBYiFYzQxDMN-_ckwddpHwC_QWPiGPgZFt_NjbX7FJxWkaAhReEALw_wcB ).
